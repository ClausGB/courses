```python
#Se importa las bibliotecas
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#datos que se van a obtener desde una url
data_url= "https://gitlab.com/dgtic5/res/-/raw/main/aprendizajeSupervizado/Yemen%20Cholera%20Outbreak%20Epidemiology%20Data%20-%20Data_Governorate_Level.csv"
data= pd.read_csv(data_url)
```

```python
#Muestra los primeros datos del dataframe
data.head()
```

```python
#Muestra información detallada del dataset
data.info()
```

# **4. Analiza y obtén las gráficas de las muertes con respecto a alguno de los códigos “COD”**

El campo "Deaths" ya es un entero, no es necesario limpiar caracteres que no son numéricos ni convertirlo a enteros

### Graficando por el código alfanumérico (COD Gov English)

```python
pivote = pd.pivot_table(data,
                        values="Deaths",
                        index = ["Date"],
                        columns = ["COD Gov English"]
                        )
```

```python
pivote
```

```python
pivote.plot(figsize=(15,10))
```

**Se observa que al tener los identificadores de país normalizados mediante el código alfanumérico, las gráficas no presentan rupturas ni cambios de color**

```python
pivote_2 = pd.pivot_table(data,
                        values="Deaths",
                        index = ["Date"],
                        columns = ["COD Gov Pcode"]
                        )
```

```python
pivote_2
```

### Graficando por el código numerico (COD Gov Pcode)

```python
pivote_2.plot(figsize=(15,10))
```

**Se observa que al tener los identificadores de país normalizados mediante el código numérico, las gráficas tampoco presentan rupturas ni cambios de color**
